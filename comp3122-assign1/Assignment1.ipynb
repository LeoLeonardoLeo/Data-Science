```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "biostats.csv"
data = np.genfromtxt(path, delimiter=',', names=True, dtype=None, encoding=None) #read text from file

age = data["Age"] #2 - fetch the age column

```

```python
print(age) #3 - print age
```

```python
#counter never changes, always 0
#everytime it loops through a new age, it adds to the total
#counter = 0
#for ages in age: #taking from age array
    #counter = counter + ages / 18 #18 because in the biostats file there are 18 different ages
#print("average age is", counter)


average = np.mean(age)
print("the average is",average)

standard_deviation = np.std(age)
print("the standard deviation is",standard_deviation)

minimum = np.min(age)
print("the minimum is", minimum)

maximum = np.max(age)
print("the maximum is", maximum)
```

```python
thirty_percentile = np.percentile(age, 30) #gets 30th percentile
median = np.median(age) #age median
seventy_percentile = np.percentile(age, 70) #70 percentile

print("30th percentile is", thirty_percentile, 
    "|| 70th percentile is", seventy_percentile,
     "|| median age is", median)
```

```python
plt.figure()
plt.hist(age, edgecolor = "black", zorder = 2)
plt.xlabel("Age") 
plt.ylabel("Number")
plt.title("Age Distribution - By Leonardo Pereira")
plt.grid()
plt.show()
```
