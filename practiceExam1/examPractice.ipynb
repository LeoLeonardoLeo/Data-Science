```python
#PLOTTING 
#using oshawa weather 2016 csv file

import matplotlib.pyplot as plt
```

```python
#we need to load file
weather = np.loadtxt("OshawaWeather2016.csv", skiprows=1, delimiter=',')
```

```python
#within the csv file there are day, max temp, min temp, team temp, total precip

#to load we can take our "weather" variable and set it to the things within the csv
#[row, column]
#we take from each column in the csv file
day = weather[:, 0] 
maxTemp = weather[:, 1] 
minTemp = weather[:, 2]   
meanTemp = weather[:, 3]  
precip = weather[:, 4] 
```

```python
#now to plot
#lets plot min temp
plt.figure(figsize=(20, 10)) # adds space

#plotting the day and min temp, linestyle means how each point it connected and color means which color(blue)
plt.plot(day, minTemp, marker="o", linestyle="-", color="b", label="Min Temp (C)" )

plt.title("Minimum temperature in degrees C")
plt.xlabel("Day")
plt.ylabel("Min Temp (C)")
plt.grid(True)
```

```python
#now lets do max temp
plt.figure(figsize=(20, 10)) # adds space
plt.plot(day, maxTemp, marker="o", linestyle="-", color="r", label="Max Temp (C)" )
plt.title("Maximum temperature in degrees C")
plt.xlabel("Day")
plt.ylabel("Max Temp (C)")
plt.grid(True)
```

```python
#mean temp
plt.figure(figsize=(20, 10)) # adds space
plt.plot(day, meanTemp, marker="o", linestyle="-", color="g", label="Meam Temp (C)" )
plt.title("Mean temperature in (C)")
plt.xlabel("Day")
plt.ylabel("Mean Temp (C)")
plt.grid(True)
```

```python
#kneighbours import
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
#loading the iris dataset
iris = datasets.load_iris()
X = iris.data  # Features (data)
y = iris.target  # Target (labels)
```

```python
#testing and training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Make an object from DecisionTreeClassifier class from sklearn.tree package
decision_tree = DecisionTreeClassifier()
```

```python
#use "fit" keyword to to pass X_train and y_train are argument  into the decision tree
decision_tree.fit(X_train, y_train)
```

```python
#make predictions now 
y_pred = decision_tree.predict(X_test)
y_pred
```

```python
#check predictions using accuracy
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
